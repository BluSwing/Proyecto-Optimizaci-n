```python
#ChatGPT
import itertools
import sys
import time

# Función para calcular la distancia total de una ruta dada
def calcular_distancia(ruta, matriz_distancias):
    distancia_total = 0
    num_nodos = len(ruta)
    for i in range(num_nodos - 1):
        nodo_actual = ruta[i]
        nodo_siguiente = ruta[i + 1]
        distancia_total += matriz_distancias[nodo_actual][nodo_siguiente]
    # Sumar la distancia de regreso al nodo inicial
    distancia_total += matriz_distancias[ruta[-1]][ruta[0]]
    return distancia_total

# Función principal para resolver el problema
def resolver_ovrp(matriz_distancias):
    num_nodos = len(matriz_distancias)
    # Generar todas las permutaciones de los nodos
    mejores_ruta = None
    mejor_distancia = sys.maxsize

    start_time = time.time()

    for ruta in itertools.permutations(range(num_nodos)):
        distancia = calcular_distancia(ruta, matriz_distancias)
        if distancia < mejor_distancia:
            mejor_distancia = distancia
            mejores_ruta = ruta

    end_time = time.time()
    tiempo_ejecucion = end_time - start_time

    return mejores_ruta, mejor_distancia, tiempo_ejecucion

# Matriz de distancias reducida a 10x10
matriz_distancias_reducida = [
    [0, 2, 9, 10, 6, 3, 8, 7, 12, 6],
    [2, 0, 4, 7, 9, 8, 5, 11, 8, 7],
    [9, 4, 0, 3, 7, 5, 10, 9, 7, 8],
    [10, 7, 3, 0, 6, 2, 8, 5, 9, 10],
    [6, 9, 7, 6, 0, 4, 3, 8, 5, 11],
    [3, 8, 5, 2, 4, 0, 6, 9, 7, 3],
    [8, 5, 10, 8, 3, 6, 0, 4, 9, 5],
    [7, 11, 9, 5, 8, 9, 4, 0, 6, 10],
    [12, 8, 7, 9, 5, 7, 9, 6, 0, 4],
    [6, 7, 8, 10, 11, 3, 5, 10, 4, 0],
]

# Resolver el problema con la matriz de distancias reducida
mejores_ruta, mejor_distancia, tiempo_ejecucion = resolver_ovrp(matriz_distancias_reducida)

# Imprimir resultados
print(f"Mejor ruta encontrada: {mejores_ruta}")
print(f"Distancia mínima: {mejor_distancia}")
print(f"Tiempo de ejecución: {tiempo_ejecucion:.6f} segundos")
```

```python
import gurobipy as gp
from gurobipy import GRB
import time

# Matriz de distancias reducida a 10x10
matriz_distancias_reducida = [
    [0, 2, 9, 10, 6, 3, 8, 7, 12, 6],
    [2, 0, 4, 7, 9, 8, 5, 11, 8, 7],
    [9, 4, 0, 3, 7, 5, 10, 9, 7, 8],
    [10, 7, 3, 0, 6, 2, 8, 5, 9, 10],
    [6, 9, 7, 6, 0, 4, 3, 8, 5, 11],
    [3, 8, 5, 2, 4, 0, 6, 9, 7, 3],
    [8, 5, 10, 8, 3, 6, 0, 4, 9, 5],
    [7, 11, 9, 5, 8, 9, 4, 0, 6, 10],
    [12, 8, 7, 9, 5, 7, 9, 6, 0, 4],
    [6, 7, 8, 10, 11, 3, 5, 10, 4, 0],
]

# Función para resolver el OVRP usando Gurobi
def resolver_ovrp_gurobi(matriz_distancias):
    num_nodos = len(matriz_distancias)

    # Crear el modelo
    m = gp.Model()

    # Crear variables
    x = m.addVars(num_nodos, num_nodos, vtype=GRB.BINARY, name="x")
    u = m.addVars(num_nodos, vtype=GRB.INTEGER, name="u")

    # Establecer la función objetivo
    m.setObjective(gp.quicksum(matriz_distancias[i][j] * x[i, j] for i in range(num_nodos) for j in range(num_nodos) if i != j), GRB.MINIMIZE)

    # Restricciones
    for i in range(num_nodos):
        m.addConstr(gp.quicksum(x[i, j] for j in range(num_nodos) if i != j) == 1)
        m.addConstr(gp.quicksum(x[j, i] for j in range(num_nodos) if i != j) == 1)

    for i in range(1, num_nodos):
        for j in range(1, num_nodos):
            if i != j:
                m.addConstr(u[i] - u[j] + num_nodos * x[i, j] <= num_nodos - 1)

    # No permitir bucles (u_i >= 2 para todos los nodos excepto el nodo de inicio)
    for i in range(1, num_nodos):
        m.addConstr(u[i] >= 2)

    # Optimizar el modelo
    m.optimize()

    if m.status == GRB.OPTIMAL:
        ruta = []
        for i in range(num_nodos):
            for j in range(num_nodos):
                if x[i, j].x > 0.5:
                    ruta.append((i, j))
        return ruta, m.objVal
    else:
        return None, None

start_time = time.time()
ruta, distancia = resolver_ovrp_gurobi(matriz_distancias_reducida)
end_time = time.time()

# Imprimir resultados
if ruta:
    print("Mejor ruta encontrada:")
    print(ruta)
    print(f"Distancia mínima: {distancia}")
else:
    print("No se encontró una solución óptima.")

print(f"Tiempo de ejecución: {end_time - start_time:.6f} segundos")

```
